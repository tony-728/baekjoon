```python
# ÌÅê2
import sys
from collections import deque


def push(x):
    global my_queue
    my_queue.append(x)


def pop():
    global my_queue
    if my_queue:
        print(my_queue.popleft())
    else:
        print(-1)


def size():
    global my_queue
    print(len(my_queue))


def empty():
    global my_queue
    if my_queue:
        print(0)
    else:
        print(1)


def front():
    global my_queue
    if my_queue:
        print(my_queue[0])
    else:
        print(-1)


def back():
    global my_queue
    if my_queue:
        print(my_queue[-1])
    else:
        print(-1)


my_queue = deque()

N = int(input())

for _ in range(N):
    command = sys.stdin.readline().split()
    if command[0] == "push":
        push(command[1])
    elif command[0] == "front":
        front()
    elif command[0] == "back":
        back()
    elif command[0] == "size":
        size()
    elif command[0] == "empty":
        empty()
    elif command[0] == "pop":
        pop()
```
