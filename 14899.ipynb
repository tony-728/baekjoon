```python
# 스타트와 링크

import sys
from itertools import permutations


def dfs(depth: int, start: int, n: int, m: int) -> int:
    """
    N/2명으로 이루어진 스타트 팀과 링크 팀으로 사람들을 나눠야 한다.
    능력치 Sij는 i번 사람과 j번 사람이 같은 팀에 속했을 때, 팀에 더해지는 능력치이다.
    팀의 능력치는 팀에 속한 모든 쌍의 능력치 Sij의 합이다. Sij는 Sji와 다를 수도 있으며,
    i번 사람과 j번 사람이 같은 팀에 속했을 때, 팀에 더해지는 능력치는 Sij와 Sji이다.

    두개의 팀 능력치의 차이의 최소값을 찾으라

    Parameters
    ----------
    depth : int
        m개의 수열까지 찾기 위한 깊이를 나타내는 파라미터
    start : int
        오름차순으로만 팀을 나누기 위한 파라미터
        - Sij + Sji로 능력치 값을 더하기 때문에 중복이 된다.
    n : int
        0부터 n-1까지 수를 나타내기 위한 n
    m : int
        N/2명, m = N/2 개의 수열을 나타냄

    Returns
    -------
    int
        두 팀의 능력치의 최소 값
    """
    global min_val

    if depth == m:
        s_hap = 0
        l_hap = 0
        # answer로 정한 팀을 제외한 나머지 팀을 구하기 위해 전체 집합에서 차집합으로 구함
        diff_list = list(n_list - set(answer))

        # 스타팀과 링크 팀의 능력치를 구하기 위한 순열
        s_list = list(permutations(answer, 2))
        l_list = list(permutations(diff_list, 2))

        # 두 팀의 능력치 각각의 합
        for s in s_list:
            s_hap += p[s[0]][s[1]]

        for l in l_list:
            l_hap += p[l[0]][l[1]]
        
        diff_sl = abs(s_hap - l_hap)

        if min_val == 0:
            min_val = diff_sl
        elif min_val > diff_sl:
            min_val = diff_sl
        elif diff_sl == 0:
            min_val = diff_sl

        return

    # 오름차순으로 팀을 구성하기 위해 start부터 반복한다.
    for i in range(start, n):
        if not visited[i]:
            visited[i] = True
            answer.append(i)
            # 깊이 탐색 시에 start를 증가하므로써 오름차순으로 팀을 구성할 수 있다.
            dfs(depth=depth + 1, start=i + 1, n=n, m=m)
            visited[i] = False
            answer.pop()

            # 두 팀의 능력치의 차이가 0보다 작을 수 없으므로 반복문 탈출
            if min_val == 0:
                break

    return min_val


n = int(input())

n_list = set(i for i in range(n))

p = []

for _ in range(n):
    p.append(list(map(int, sys.stdin.readline().split())))

visited = [False] * (n + 1)
answer = []
min_val = 0

print(dfs(0, 0, n, n // 2))
```

```python
from itertools import combinations
import sys
input = sys.stdin.readline


N = int(input())
members = [i for i in range(N)]
synerge = []
min_val = 1e9

for i in range(N):
    synerge.append(list(map(int, input().split())))

team_combs = list(combinations(members, N//2))


for i in range(len(team_combs)//2):
    team1 = list(team_combs[i])
    team2 = list(team_combs[-1-i])
    t1_total = 0
    t2_total = 0

    for i in range(N//2):
            for j in range(N//2):
                if i == j:
                    continue
                t1_total += synerge[team1[i]][team1[j]]
                t2_total += synerge[team2[i]][team2[j]]

    offset = t1_total - t2_total
    if offset < 0:
        offset *= -1
    if offset < min_val:
        min_val = offset

print(min_val)
```
