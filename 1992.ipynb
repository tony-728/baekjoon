```python
# 쿼드트리
import sys

input = sys.stdin.readline


def quad_tree(n: int, x: int, y: int):
    """
    모두 0으로만 되어 있으면 압축 결과는 "0"이 되고, 모두 1로만 되어 있으면 압축 결과는 "1"이 된다.
    만약 0과 1이 섞여 있으면 전체를 한 번에 나타내지를 못하고,
    왼쪽 위, 오른쪽 위, 왼쪽 아래, 오른쪽 아래, 이렇게 4개의 영상으로 나누어 압축하게 되며,
    이 4개의 영역을 압축한 결과를 차례대로 괄호 안에 묶어서 표현한다

    Parameters
    ----------
    n : int
        영상을 확인하는 길이
    x : int
        x 좌표
    y : int
        y 좌표
    """
    # 길이가 1일 때 확인 후 리턴
    if n == 1:
        print(image[y][x], end="")
        return

    # 확인하는 정사각형이 모두 같은 숫자로 되어있는지 확인한다.
    flag = True
    for i in range(y, y + n):
        if not flag:
            break
        for j in range(x, x + n):
            if image[y][x] != image[i][j]:
                flag = False
                break

    # 모두 같은 숫자로 되어 있을 때
    if flag:
        print(image[y][x], end="")
    # 하나 이상이 다른 숫자로 되어 있을 때
    else:
        # 확인하는 길이를 반으로 줄인다.
        decreased_n = n // 2
        print("(", end="")
        # 확인하는 길이를 반으로 줄였다.
        # 가로, 세로가 반으로 줄었으므로 결과적으로 4등분되어 확인한다.
        quad_tree(decreased_n, x, y)
        quad_tree(decreased_n, x + decreased_n, y)
        quad_tree(decreased_n, x, y + decreased_n)
        quad_tree(decreased_n, x + decreased_n, y + decreased_n)
        print(")", end="")


n = int(input())
image = [list(input().rstrip()) for _ in range(n)]

quad_tree(n, 0, 0)
```
