```python
# 종이의 개수

import sys
'''
N×N크기의 행렬로 표현되는 종이가 있다. 종이의 각 칸에는 -1, 0, 1 중 하나가 저장되어 있다.
우리는 이 행렬을 다음과 같은 규칙에 따라 적절한 크기로 자르려고 한다.

만약 종이가 모두 같은 수로 되어 있다면 이 종이를 그대로 사용한다.
(1)이 아닌 경우에는 종이를 같은 크기의 종이 9개로 자르고, 각각의 잘린 종이에 대해서 (1)의 과정을 반복한다.
이와 같이 종이를 잘랐을 때, -1로만 채워진 종이의 개수, 0으로만 채워진 종이의 개수, 1로만 채워진 종이의 개수를 구해내는 프로그램을 작성하시오.
'''
input = sys.stdin.readline


def decompose(n: int, x: int, y: int):
    """
    모두 같은 숫자로 채워져 있는지 확인한다.

    Parameters
    ----------
    n : int
        확인하는 길이
    x : int
        x 좌표
    y : int
        y 좌표
    """
    global one, zero, m_one
    # 길이가 1일 때 어떤 숫자인지 확인 후 리턴
    if n == 1:
        if image[y][x] == "1":
            one += 1
        elif image[y][x] == "0":
            zero += 1
        else:
            m_one += 1

        return

    # 길이 n만큼에서 모두 같은 숫자로 이뤄졌는지 확인한다.
    flag = True
    for i in range(y, y + n):
        if not flag:
            break
        for j in range(x, x + n):
            if image[y][x] != image[i][j]:
                flag = False
                break
    
    # 모두 같은 숫자로 이뤄짐
    if flag:
        if image[y][x] == "1":
            one += 1
        elif image[y][x] == "0":
            zero += 1
        else:
            m_one += 1
    # 하나 이상의 숫자가 다름
    else:
        # 종이를 9개로 나누기 위해 3으로 나눈다.
        decreased_n = n // 3

        # 나눈 가로 세로 만큼 다시 확인한다.
        # 가로, 세로를 3으로 나눴으므로 총 9개
        decompose(decreased_n, x, y)

        decompose(decreased_n, x + decreased_n, y)
        decompose(decreased_n, x + (decreased_n * 2), y)
        decompose(decreased_n, x + (decreased_n * 2), y + decreased_n)

        decompose(decreased_n, x, y + decreased_n)
        decompose(decreased_n, x, y + (decreased_n * 2))
        decompose(decreased_n, x + decreased_n, y + (decreased_n * 2))

        decompose(decreased_n, x + decreased_n, y + decreased_n)
        decompose(decreased_n, x + (decreased_n * 2), y + (decreased_n * 2))


one = 0
zero = 0
m_one = 0

n = int(input())
image = [list(input().split()) for _ in range(n)]

decompose(n, 0, 0)

print(m_one)
print(zero)
print(one)
```
