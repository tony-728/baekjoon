```python
# N과 M (3)
# 1부터 N까지 자연수 중에서 M개를 고른 수열
# 같은 수를 여러번 골라도 된다.

n, m = map(int, input().split())

answer = []

def n_and_m(depth:int, n:int, m:int):
    """
    자연수 n과 m이 주어였을 때 1부터 n까지 자연수 중에서 m개를 고른 수열
    중복 가능

    Parameters
    ----------
    depth : int
        찾는 깊이 설정
    n : int
        1부터 n까지 수열을 만들기 위한 파라미터
    m : int
        m개 수열을 
    """
    if depth == m:
        print(*answer)
        return

    else:
        for i in range(1, n+1):
            answer.append(i)
            n_and_m(depth=depth+1, n=n, m=m)
            answer.pop()

n_and_m(0, n, m)
```

```python
n, m = list(map(int,input().split()))
s = []

def dfs():
    if len(s) == m:
        print(' '.join(map(str, s)))
        return
    for i in range(1, n+1):
        s.append(i)
        dfs()
        s.pop()
dfs()
```
