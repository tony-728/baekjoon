```python
# 색종이 만들기
"""
전체 종이가 모두 같은 색으로 칠해져 있지 않으면 가로와 세로로 중간 부분을 잘라서 
<그림 2>의 I, II, III, IV와 같이 똑같은 크기의 네 개의 N/2 × N/2색종이로 나눈다.
나누어진 종이 I, II, III, IV 각각에 대해서도 앞에서와 마찬가지로 모두 같은 색으로 칠해져 있지 않으면
같은 방법으로 똑같은 크기의 네 개의 색종이로 나눈다.

이와 같은 과정을 잘라진 종이가 모두 하얀색 또는 모두 파란색으로 칠해져 있거나,
하나의 정사각형 칸이 되어 더 이상 자를 수 없을 때까지 반복한다.
"""

import sys

input = sys.stdin.readline


def decompose(n: int, x: int, y: int):
    """
    색종이 만들기 분할 정복
    정사각형에 모두 같은 색상(숫자)로 되어있는지 확인한다.
    하나라도 다른 색상이 있으면 4등분하여 다시 확인한다.
    최종적으로 길이가 1일때까지 수행한다.

    Parameters
    ----------
    n : int
        색상을 확인하는 정사각형의 길이
    x : int
        시작 x 좌표
    y : int
        시작 y 좌표
    """

    global w, b

    # 길이가 1일때 색상을 확인한다.
    if n == 1:
        if image[y][x] == "0":
            w += 1
        else:
            b += 1

        return

    # 정사각형에 다른 숫자가 있는지 확인한다.
    flag = True
    for i in range(y, y + n): # y축 
        if not flag:
            break
        for j in range(x, x + n): # x축
            # 정사각형의 첫번째 색상과 같은지 확인
            # 다르다면 flag를 바꾼다.
            if image[i][j] != image[y][x]: 
                flag = False
                break

    # 모든 색상이 같다.
    # 어떤 색상인지 확인 후 카운트 증가
    if flag:
        if image[y][x] == "0":
            w += 1
        else:
            b += 1
    # 한개의 색상도 다른 것이 존재
    else:
        # 확인하는 길이를 반으로 줄인다.
        decreased_n = n // 2

        # 정사각형을 4등분하여 확인한다.
        # 확인하는 길이와 x, y좌표에 확인하는 길이만큼 더하여 4등분을 확인
        decompose(decreased_n, x, y)
        decompose(decreased_n, x + decreased_n, y)
        decompose(decreased_n, x, y + decreased_n)
        decompose(decreased_n, x + decreased_n, y + decreased_n)


w = 0
b = 0

n = int(input())
image = [list(input().split()) for _ in range(n)]

decompose(n, 0, 0)

print(w)
print(b)
```
