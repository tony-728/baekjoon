```python
# 행렬 제곱
"""
크기가 N*N인 행렬 A가 주어진다.
이때, A의 B제곱을 구하는 프로그램을 작성하시오.
수가 매우 커질 수 있으니, A^B의 각 원소를 1,000으로 나눈 나머지를 출력한다.
"""
import sys

input = sys.stdin.readline


def mat_multi(matrix_a: list, matrix_b: list) -> list:
    """
    행렬 곱셈

    Parameters
    ----------
    matrix_a : list
        행렬
    matrix_b : list
        행렬

    Returns
    -------
    list
        곱셈된 행렬
    """
    answer = [n_a * [0] for i in range(n_a)]
    # n_a만큼 반복
    for i in range(n_a):
        # m_b만큼 반복
        # 그래야 n_a x m_b크기의 행렬이 만들어 진다.
        for j in range(n_a):
            # n_b만큼 반복하므로써 실제 행렬의 곱을 수행
            for k in range(n_a):
                answer[i][j] += matrix_a[i][k] * matrix_b[k][j]
            # 행렬의 곱셈을 수행하는 동안 수가 매우 커질 수 있으므로 문제에서
            # 1000을 나눈 나머지로 대신함
            answer[i][j] = answer[i][j] % 1000

    return answer


def decompose(n:int, metrix:list)->list:
    """
    행렬의 거듭제곱 수행

    Parameters
    ----------
    n : int
        거듭제곱 수행 횟수
    metrix : list
        행렬

    Returns
    -------
    list
        n거듭제곱된 행렬
    """
    # 재귀함수의 base조건
    if n == 1:
        for x in range(len(metrix)):
            for y in range(len(metrix)):
                # 문제에서 각 원소는 1000보자 작거나 같은 자연수라고
                # 했으므로 1000을 나눈 나머지로 각 원소를 대신함
                metrix[x][y] %= 1000
        return metrix

    else:
        # 거듭제곱 횟수를 2로 나눠주어 총 계산 횟수를 줄일 수 있다.
        m = decompose(n // 2, metrix)
        # 거듭제곱 수행
        temp = mat_multi(m, m)

        # n이 짝수면 거듭제곱 수행한 값을 리턴
        # decompose함수를 리턴
        if n % 2 == 0:
            return temp
        # n이 홀수면 원래 행렬을 한번 더 곱해야 총 거듭제곱 횟수가 달라지지 않는다.
        else:            
            return mat_multi(metrix, temp)


matrix_a = []
n_a, b = map(int, input().split())
for _ in range(n_a):
    matrix_a.append(list(map(int, input().split())))

answer = decompose(b, matrix_a)

for row in answer:
    print(*row)
```
