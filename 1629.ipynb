```python
# 곱셈
"""
자연수 A를 B번 곱한 수를 알고 싶다.
단 구하려는 수가 매우 커질 수 있으므로 이를 C로 나눈 나머지를 구하는 프로그램을 작성하시오.
"""
import sys

input = sys.stdin.readline

a, b, c = map(int, input().split())


def decompose(a: int, n: int):
    """
    자연수 a를 n번 곱한 수를 c로 나눈 나머지를 찾는다.
    계속 반씩 줄여가면서 구할 수 있는 이유는 다음과 같다.
    - 지수법칙
    - 나머지 분배 법칙
    - ex) a=10, n = 11, c = 12
    (10^11) % 12
    -> 10^11 = 10^5 x 10^5 x 10으로 표현할 수 있다.
    (10^10) % 12
    -> 10^10 = 10^5 x 10^5로 표현할 수 있다.


    Parameters
    ----------
    a : int
        곱하고자하는 수
    n : int
        곱하는 횟수

    Returns
    -------
    n이 짝수면
    -> a^n = a^n//2 x a^n//2
    n이 홀수면
    -> a^n = a^n//2 x a^n//2 x a
    """
    if n == 1:
        return a % c

    else:
        temp = decompose(a, n // 2)
        if n % 2 == 0:
            return (temp * temp) % c
        else:
            return (temp * temp * a) % c


print(decompose(a, b))
```
