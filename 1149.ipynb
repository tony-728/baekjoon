```python
# RGB거리
# 1번 2번 집은 색이 달라야한다.
# i번 집은 i-1, i+1과 색이 달라야한다.
import sys
N = int(input())

# 1. 최소값을 저장한다.
# 다음 값이 왔을 때 겹치는 자리에 더 작은 값이 오면 이전 값에서 2번째로 작은 값을 고른다.
# 만약 다음 자리에 최소값이 이전 값과 겹치는 자리이고 더 큰 값이면 다음 자리에 2번째로 작은 값을 고른다.

answer = {}
my_RGB = []
for i in range(N):
    R,G,B = map(int, sys.stdin.readline().split())
    dic = {0:R, 1:G, 2:B}
    my_RGB.append(dic)
    min_val = min(R,G,B)
    if answer:
        j = i
        while j > 0:
            # 현재의 최소값의 위치와 이전 최소값의 위치가 다르다.
            if my_RGB[j-1][answer[j-1]] != dic[min_val]:
                answer[j] = min_val
            # 현재의 최소값의 위치와 이전 최소값의 위치가 같다.
            else:
                # 이전 최소값이 현재 최소값보다 작다
                if answer[j-1] < min_val:
                    snd_min_val = sorted(dic)[1]
                    answer[j] = snd_min_val
                else:
                    snd_min_val = sorted(my_RGB[j-1])[1]
                    answer[j-1] = snd_min_val
                    answer[j] = min_val

            
            
            j -= 1

    else:
        answer[i] = min_val

print(sum(answer.values()))
```

```python
# RGB거리
# 1번 2번 집은 색이 달라야한다.
# i번 집은 i-1, i+1과 색이 달라야한다.
import sys
N = int(input())

# 1. 최소값을 저장한다.
# 다음 값이 왔을 때 겹치는 자리에 더 작은 값이 오면 이전 값에서 2번째로 작은 값을 고른다.
# 만약 다음 자리에 최소값이 이전 값과 겹치는 자리이고 더 큰 값이면 다음 자리에 2번째로 작은 값을 고른다.

answer = {}
my_RGB = []
for i in range(N):
    R,G,B = map(int, sys.stdin.readline().split())
    dic = {0:R, 1:G, 2:B}
    my_RGB.append(dic)
    sorted_dic = sorted(dic.items(), key=lambda x: x[1]) # [(index, value), (), ()k]
    if answer:
        # 현재의 최소값의 위치와 이전 최소값의 위치가 다르다.
        if answer[i-1][0] != sorted_dic[0][0]:
            answer[i] = sorted_dic[0]

        # 현재의 최소값의 위치와 이전 최소값의 위치가 같다.
        else:
            # 이전 최소값이 현재 최소값보다 작다
            if answer[i-1][1] < sorted_dic[0][1]:
                answer[i] = sorted_dic[1]

            # 이전 최소값이 현재 최소값보다 크다
            else:
                answer[i-1] = sorted(my_RGB[i-1].items(), key=lambda x:x[1])[1]
                answer[i] = sorted_dic[0]

                j = i-1
                # 이전 값 확인
                while j >= 0:
                    # 이전값끼리의 위치 확인
                    if answer[j-1][0] != answer[j][0]:
                        break
                    else:
                        answer[j-1] = sorted(my_RGB[i-1].items(), key=lambda x:x[1])[0]
                    j -= 1


    else:
        answer[i] = sorted_dic[0]

print(sum(answer.values()))
```

```python
import sys
"""
1번 집의 색은 2번 집의 색과 같지 않아야 한다.
N번 집의 색은 N-1번 집의 색과 같지 않아야 한다.
i(2<=i<=N-1)번 집의 색은 i-1번, i+1번 집의 색과 같지 않아야 한다.

solution
- 첫번째의 RGB[0][0],RGB[0][1],RGB[0][2]의 값은 각각 빨간색, 초록색, 파란색의 첫번째 값을 넣어준다.
- 2번째 값부터는 잡의 연속된 색상을 가지지 말아야하기 때문에 다음 빨간색 집에 현재 빨간색 비용과 이전 초록색, 파란색 비용중 작은 값을 더한다.
 -> RGB[i][0] = min(RGB[i-1][1], RGB[i-1][2]) + RGB[i][0]
- 빨간색, 초록색, 파란색을 동일하게 진행한다.
- 마지막 빨,초,파 중 작은 값을 출력한다.

"""

n = int(input())

RGB = []
for i in range(n):
    RGB.append(list(map(int, sys.stdin.readline().split())))

for i in range(1, len(RGB)):
    RGB[i][0] = min(RGB[i-1][1], RGB[i-1][2]) + RGB[i][0]
    RGB[i][1] = min(RGB[i-1][0], RGB[i-1][2]) + RGB[i][1]
    RGB[i][2] = min(RGB[i-1][0], RGB[i-1][1]) + RGB[i][2]

print(min(RGB[n-1][0],RGB[n-1][1],RGB[n-1][2]))
```
