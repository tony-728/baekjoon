```python
# Îç±

import sys
from collections import deque


def push_front(x):
    global my_deck
    my_deck.appendleft(x)


def push_back(x):
    global my_deck
    my_deck.append(x)


def pop_front():
    global my_deck
    if my_deck:
        print(my_deck.popleft())
    else:
        print(-1)


def pop_back():
    global my_deck
    if my_deck:
        print(my_deck.pop())
    else:
        print(-1)


def size():
    global my_deck
    print(len(my_deck))


def empty():
    global my_deck
    if my_deck:
        print(0)
    else:
        print(1)


def front():
    global my_deck
    if my_deck:
        print(my_deck[0])
    else:
        print(-1)


def back():
    global my_deck
    if my_deck:
        print(my_deck[-1])
    else:
        print(-1)


my_deck = deque()

N = int(input())

for _ in range(N):
    command = sys.stdin.readline().split()
    if len(command) == 1:
        if command[0] == "pop_front":
            pop_front()
        elif command[0] == "pop_back":
            pop_back()
        elif command[0] == "size":
            size()
        elif command[0] == "empty":
            empty()
        elif command[0] == "front":
            front()
        elif command[0] == "back":
            back()
    else:
        if command[0] == "push_front":
            push_front(command[1])
        if command[0] == "push_back":
            push_back(command[1])
```
